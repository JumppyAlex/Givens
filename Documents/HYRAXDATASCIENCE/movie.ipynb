```python
#http://192.168.22.249/files/
import pandas as pd
data = pd.read_csv('data/movies.csv')
data.tail(50)
```

```python
#Step 1: Lets see which movie has highest rating on avg
data.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
#Step 2 : How many people rated the movie
data.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
#Combine mean and count 
ratings = pd.DataFrame(data.groupby('title')['rating'].mean())
ratings['no_of_ratings'] = data.groupby('title')['rating'].count()
ratings.head(50)
```

```python
#Do pivot Table ,Pivot takes 3 variables
pivot = data.pivot_table(index='user_Id',columns='title',values='rating')
pivot
#Shows which users , what movie they rated ,what rating they gave
```

```python
#Pick a movie
movie = pivot['Liar Liar (1997)']

#Associate above and relate with others in the pivot
similar = pivot.corrwith(movie)
similar
#Put similar in a DataFrame
df = pd.DataFrame(similar,columns=['Correlations'])
df.sort_values('Correlations' ,ascending= False).head(20)

#At this point we can see the movie similar to Liar Liar
#Question how many people rated below movies
df = df.join(ratings['no_of_ratings'])
df[df['no_of_ratings']>100].sort_values('Correlations',
                                        ascending=False).head(20)
```

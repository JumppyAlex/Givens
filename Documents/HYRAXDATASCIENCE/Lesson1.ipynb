```python
import pandas as pd #To read and prepare data 
import  matplotlib.pyplot as plt #Plotting
import seaborn as sns #Plotting
```

```python
data =pd.read_csv('data/bank.csv')
data
```

```python
#Lets get bsic statistics
data.describe()
```

```python
#Plots - visualization
#Distribution of Gender
sns.countplot(x='Gender',data=data,palette='Reds')
#Blues, Reds,
```

```python
#Matplotlib
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-dark-palette')
sns.countplot(x='Education',data=data,palette='Reds')
```

```python
plt.style.use( 'seaborn-v0_8-darkgrid')
sns.countplot(x='Self_Employed',data=data,palette='Blues')
```

```python
plt.style.use( 'seaborn-v0_8-darkgrid')
sns.countplot(x='Property_Area',data=data,palette='Reds')
```

```python
plt.style.use( 'seaborn-v0_8-darkgrid')
sns.countplot(x='Loan_Status',data=data,palette='Blues')
```

```python
#Lets see the average loan
plt.style.use('bmh')
sns.barplot(x='Gender',y='LoanAmount',data=data)
plt.ylabel('Avg.Loan  Amount- USD')
plt.xlabel('Gender Category')
plt.title('Avg Loan Amount by Gender')
```

```python
plt.style.use('bmh')
sns.barplot(x='Education',y='LoanAmount',data=data)
plt.ylabel('Avg.Loan  Amount- USD')
plt.xlabel('Education')
plt.title('Avg Loan Amount by Education')
```

```python
plt.style.use('bmh')
sns.barplot(x='Education',y='LoanAmount',data=data ,hue='Gender',palette='Blues')
plt.ylabel('Avg.Loan  Amount- USD')
plt.xlabel('Education  Category')
plt.title('Avg Loan Amount by Education, Gender')
```

```python
plt.style.use('bmh')
sns.barplot(x='Education',y='ApplicantIncome',data=data ,hue='Gender',palette='Blues')
plt.ylabel('Avg Applicant Income- USD')
plt.xlabel('Education  Category')
plt.title('Avg Loan Amount by Education, Gender')
```

```python
#Swarm plot
sns.swarmplot(x='Gender',y='LoanAmount',data=data,palette='Reds' ,hue='Education')
```

```python
#Pie chart
plt.style.use('ggplot')
data.groupby('Gender').size().plot(kind='pie',autopct="%1.1f")
#Above %1.1 means show iin 1 decimal place.
```

```python
#Pie chart
plt.style.use('ggplot')
data.groupby('Education').size().plot(kind='pie',autopct="%1.1f")
#Above %1.1 means show iin 1 decimal place.
```

```python
#Pie chart
plt.style.use('ggplot')
data.groupby('Property_Area').size().plot(kind='pie',autopct="%1.1f")
#Above %1.1 means show iin 1 decimal place.
```

```python
import pandas as pd
data = pd.read_csv('data/Mall_Customers.csv')
data.dropna(inplace = True) # removes empty records
data
```

```python
# step1 - split data to x
array = data.values
X = array[:, 2:5] # age,income,spending score
```

```python
# step2 - fit the model
from sklearn.cluster import KMeans
model = KMeans(n_clusters=10, random_state=42)
model.fit(X) # we fit 100 rows 
```

```python
# step3 - get the clusters
means = model.cluster_centers_
cluster = pd.DataFrame(means, columns = ['Age','Income','Score'])
cluster
```

```python
# group2 - Avg age- 43, Avg income - 106k, Avg score - 19
# 1.They earn very well
# 2.They might be parents
# 3.They have a low shopping experience
# The mall should get their details, reach out, understand their needs
# The mall could bring more items, clothes, furniture
# Get group 2 members
data['label'] = model.labels_
segment = data[data['label'] == 2]
segment 
#coding.co.ke/datasets
#EastWestAirline.csv
#Save it in ur hyrax data folder
```

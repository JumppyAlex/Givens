```python
import pandas as pd
data = pd.read_csv('data/EastWestAirlines.csv')
data
```

```python
array = data.values
X = array[:,0:4]
Y = array[:,4]
Y
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=10, random_state=42)
model.fit(X) # we fit 100 rows 
```

```python
means = model.cluster_centers_
cluster = pd.DataFrame(means,columns =['Flight_miles','Flight_trans','Day_since_enroll','Balance'])
cluster
```

```python
data['label'] = model.labels_
segment = data[data['label'] == 2]
segment 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
data = pd.read_csv ('data/Automobile_data.csv')
data
```

```python

data['make'].replace({'alfa-romero':50, 'audi':51, 'volvo':52, 'subaru':53, 'toyota':54, 'saab':55,
                       'mercury':56, 'volkswagen':57, 'renault':58,  'porsche':59, 
                       'plymouth':60, 'peugot':61, 'nissan':62, 'mitsubishi':63,
                        'mercedes-benz':64, 'mazda':65, 'jaguar':66, 'isuzu':67,
                        'honda':68, 'dodge':69, 'chevrolet':70, 'bmw':71}, inplace=True)
data.groupby("make").size()
```

```python
data.groupby("drive-wheels").size()
data['drive-wheels'].replace({'rwd':45, '4wd':46, 'fwd':47}, inplace=True)
```

```python
data.groupby("body-style").size()
data['body-style'].replace({'sedan':35, 'hatchback':36, 'convertible':37, 'wagon':38, 'hardtop':39}, inplace=True)
```

```python
data.groupby("num-of-doors").size()
data['num-of-doors'].replace({'two':31, 'four':30}, inplace=True)
```

```python
data.groupby("num-of-cylinders").size()
data['num-of-cylinders'].replace({'two':1, 'four':2, 'five':3, 'six':4, 'eight':5, 'twelve':6}, inplace=True)
```

```python
data.groupby("fuel-type").size()
data['fuel-type'].replace({'gas':25, 'diesel':24}, inplace=True)
```

```python
data.groupby("aspiration").size()
data['aspiration'].replace({'std':20, 'turbo':21}, inplace=True)
```

```python
data.groupby("fuel-system").size()
data['fuel-system'].replace({'mpfi':0, 'idi':1, '2bbl':2, '1bbl':3, '4bbl':4, 'spdi':5, 'mfi':6, 'spfi':7}, inplace=True)
```

```python
data.groupby("engine-type").size()
data['engine-type'].replace({'dohc':0, 'dohcv':1, 'l':2, 'ohc':3, 'ohcf':4, 'ohcv':5, 'rotor':6}, inplace=True)
```

```python
data.groupby("make").size()
data['make'].replace({'audi':9, 'volvo':8, 'alfa-romero':7}, inplace=True)
```

```python
data.groupby("engine-location").size()
data['engine-location'].replace({'front':11, 'rear':12}, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
x =LabelEncoder()
data ['body-style'] = x.fit_transform(data['body-style'])
data.to_csv('toyota.csv')
```

```python
array=data.values
x=array[:,0:25 ]
y=array[:, 25]
y
```

```python
import sklearn
from sklearn import model_selection
#Step 2: Split all the 2105 rows into 2 parts
#Parts 1 wil have training data 70% data
#Part 2 will have testing data 30% data


x_train ,x_test,y_train,y_test=model_selection.train_test_split(x,y,
                                                        test_size=0.3,
                                                        random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC # Support Vector Machine ,SVM

#Switch the models
model=LinearRegression()
#Fit only 70% of data
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
print('Answers',predictions)
print('Expected',y_test)
#The model predicts the 30% of data we compare results with Y_test
from sklearn.metrics  import r2_score
print(r2_score(predictions,y_test))
```
